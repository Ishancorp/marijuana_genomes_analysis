# Data input

```python
import pandas as pd
import numpy as np
import bq_helper
from bq_helper import BigQueryHelper
# https://www.kaggle.com/sohier/introduction-to-the-bq-helper-package
genomes = bq_helper.BigQueryHelper(active_project="bigquery-public-data",
                                   dataset_name="genomics_cannabis")
```

```python
bq_assistant = BigQueryHelper("bigquery-public-data", "genomics_cannabis")
bq_assistant.list_tables()
```

```python
bq_assistant.table_schema("cs3k_project_info")
```

```python
bq_assistant.head("MNPR01_201703", num_rows=10)
```

```python
query_top_ten = """SELECT
  *
FROM
  `bigquery-public-data.genomics_cannabis.MNPR01_201703` v
LIMIT 10;"""
response_top_ten = genomes.query_to_pandas_safe(query_top_ten, max_gb_scanned=100)
response_top_ten.columns
```

```python
response_top_ten
```

```python
query_reduced = """SELECT
  variant_id, alternate_bases, quality, RPPR, type, len, cigar
FROM
  `bigquery-public-data.genomics_cannabis.MNPR01_201703` v
limit 10;"""
response_reduced = genomes.query_to_pandas_safe(query_reduced, max_gb_scanned=100)
response_reduced.head(10)
```

alternate_bases consists of genes of the variant, while type means type of genetic variation. All the various random letter combinations represent proteins. 

Proteins are: 'AB', 'ABP', 'AC', 'AF', 'AN', 'AO', 'DP', 'DPB', 'DPRA', 'EPP', 'EPPR', 'GTI', 'MQM', 'MQMR', 'NS', 'PAO', 'PQA', 'PQR', 'PRO', 'QA', 'QR', 'RO', 'RPL', 'RPP', 'RPPR', 'RPR', 'RUN', 'SAF', 'SAP', 'SAR', 'SRF', 'SRP', 'SRR'
